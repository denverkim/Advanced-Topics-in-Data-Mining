```python
# Telco Churn 데이터 전처리 + 의사결정나무 모델 코드
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score



# 2. TotalCharges를 수치형으로 변환 (결측치는 NaN 처리됨)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")








# 8. 의사결정나무 모델 학습
clf = DecisionTreeClassifier(random_state=42, max_depth=5)
clf.fit(X_train, y_train)

# 9. 예측 및 정확도
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("의사결정나무 정확도:", accuracy)



# 11. Feature Importance 시각화
importances = clf.feature_importances_
feat_importances = pd.Series(importances, index=X.columns)
feat_importances.sort_values(ascending=False).plot(kind="bar", figsize=(12,5))
plt.title("Feature Importance")
plt.show()
```

```python
# 교차검증
from sklearn.model_selection import cross_val_score
lm_cv = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
print(lm_cv)
print(lm_cv.mean(), lm_cv.std())
```
